```
#importing library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
#import the dataset
from google.colab import files 


uploaded = files.upload()

import io
ds = pd.read_csv(io.BytesIO(uploaded['Cancer_dataset.csv'])) 
print(ds)
```

```
#extracting independent values
X = ds.iloc[:,2:-1].values
print(X)
print(X.shape)
```

```
#extracting dependent values
Y = ds.iloc[:,1:2].values
print(Y)
print(Y.shape)
```

```
#plotting graph between first 5 features of dataset
sns.pairplot(ds, hue='diagnosis', vars=  ["radius_mean", "texture_mean", "perimeter_mean", "area_mean", "smoothness_mean"])
```

```
#plotting graph between last 5 features of dataset
sns.pairplot(ds, hue='diagnosis', vars=  ["compactness_mean","concavity_mean","concave points_mean","symmetry_mean","fractal_dimension_mean"])
```

```
plt.figure(figsize=(20, 10))
sns.heatmap(ds.corr(), annot= True)
```

```
#splitting the data set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=10)
```

```
print(X_train)
```

```
len(X_train)
```

```
print(X_test)
```

```
len(X_test)
```

```
#feature scalling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
print(X_train)
print(X_test)
```

```
#fit the SVM model to training dataset with kernel=linear
from sklearn.svm import SVC
svm_model = SVC(kernel='linear')
svm_model.fit(X_train,Y_train)
```

```
#prediction on test dataset
prediction = svm_model.predict(X_test)
print(prediction)
```

```
#confusion matrix
from sklearn import metrics
cm = metrics.confusion_matrix(Y_test,prediction)
print(cm)
```

```
#heatmap
sns.heatmap(cm,annot=True)
plt.title('Confusion Matrix')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.show()
```

```
#printing accuracy
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_test,prediction)*100
print(accuracy)
```

```
from sklearn.metrics import classification_report
print(classification_report(Y_test,prediction))
```

```
#fit the SVM model to training dataset with kernel = rbf
svm_rbf_model = SVC(kernel='rbf')
svm_rbf_model.fit(X_train,Y_train)
```

```
#Prediction
pred = svm_rbf_model.predict(X_test)
print(pred)
```

```
#confusion matrix
cm = metrics.confusion_matrix(Y_test,pred)
print(cm)
```

```
#heatmap
sns.heatmap(cm,annot=True)
plt.title('Confusion matrix')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.show()
```

```
#accuracy
acc = accuracy_score(Y_test,pred)*100
print(acc)
```

```
print(classification_report(Y_test,pred))
```
